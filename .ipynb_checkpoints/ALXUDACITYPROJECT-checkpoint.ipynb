#Significant of research question(Distribution of Firarms per states between 2010 -2016)
This findings enable us to know the distribution mean, standard deviation, highest permit of firearms per state etc

It is discover that Kentucky state has the highest demand of firearms  between 2010-2016

```python
#Extracting dataset between 2010-01-01 & 2016-01-01
df_permit_check20102016 = df_permit_check.loc[(df_permit_check["month"] >= "2010-01-01") & (df_permit_check["month"] <= "2016-12-01")]

#df_permit_check20102016
```

```python
#Identity the total firearms per state between 2010-2016
df_permit_check20102016A = df_permit_check20102016[['state', 'totals']]
```

```python
#grouping and renaming
re_grouped= df_permit_check20102016A.groupby('state').agg({'totals': ['sum', 'min', 'max', 'std']})
re_grouped.columns = ['sum','min', 'max', 'std']
re_grouped = re_grouped.reset_index(drop=False)
re_grouped.head()

#Source:https://jamesrledoux.com/code/group-by-aggregate-pandas

```

```python
#state with the highest total number between 2010 -2016
re_grouped[re_grouped['sum'] == re_grouped['sum'].max()]
```

```python
# Plotting the pie chart for above 50 states between 2010-2016
re_grouped.groupby(['state']).sum().plot(kind='pie', y='sum', rot= 1090, title="population distribution between 2010-2016", figsize=(25,25))

plt.legend(loc='upper left', title='States', labelspacing=0.1)
```

### Research Question 2  (Visuals and Distribution!)

Histogram showing total firearms per state as at 2010 & 2016. Result shos both are rightly skewed

```python
def show_distribution(dataset, title):
    plt.hist(dataset['totals'], bins=20, alpha=0.5, histtype='stepfilled', color='steelblue', edgecolor='none')

    plt.xlabel('total')
    plt.ylabel('Counts')
    plt.title(title)
```

```python
show_distribution(gun_diff2016, 'Total firearm distribution_2016')
```

```python
#gun_diff['totals'].hist(bins = 20)

show_distribution(gun_diff, 'Total firearm distribution_2010')
```
